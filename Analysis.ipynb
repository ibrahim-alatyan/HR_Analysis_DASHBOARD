```python
#Q1/ How many total employees are there?
total_employees_py = len(df)
print(f"total of employees is: {total_employees_py}")
```

```python
#Q2/ What is the employee count for each department?
dept_counts_py = (df.groupby("Department")
               .size()
               .reset_index(name = "count")
)
dept_counts_py
```

```python
#Q3/ What is the average monthly income by job role?
avg_income_byJob_py = (df.groupby("JobRole")["MonthlyIncome"]
                    .mean()
                    .reset_index(name = "AVG INCOME")
)
avg_income_byJob_py
```

```python
#Q4/ Who are the top 5 employees by performance rating?
top5_Performance_py = df.sort_values(["PerformanceRating"] , ascending=False).head(5)
top5_Performance_py
```

```python
#Q5/ Which department has the highest average performance rating?
highAVG_dep_py = (df.groupby("Department")["PerformanceRating"]
               .mean()
               .reset_index(name = "AvgPerformance")
               .sort_values("AvgPerformance", ascending=False)
               .head(1)
)
highAVG_dep_py
```

```python
#Q6/ What Average age of employees for each Department?
avgAGE_Emp_py = (df.groupby("Department")["Age"]
              .mean()
              .reset_index(name = "AVG_Age")
)
avgAGE_Emp_py
```
