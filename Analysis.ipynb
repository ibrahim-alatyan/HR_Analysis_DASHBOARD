```python
#Q1/ How many total employees are there?
total_employees_py = len(df)
print(f"total of employees is: {total_employees_py}")
```

```python
#Q2/ What is the employee count for each department?
dept_counts_py = (df.groupby("Department")
               .size()
               .reset_index(name = "count")
)
dept_counts_py
```

```python
#Q3/ What is the average monthly income by job role?
avg_income_byJob_py = (df.groupby("JobRole")["MonthlyIncome"]
                    .mean()
                    .reset_index(name = "AVG INCOME")
)
avg_income_byJob_py
```

```python
#Q4/ Who are the top 5 employees by performance rating?
top5_Performance_py = df.sort_values(["PerformanceRating"] , ascending=False).head(5)
top5_Performance_py
```

```python
#Q5/ Which department has the highest average performance rating?
highAVG_dep_py = (df.groupby("Department")["PerformanceRating"]
               .mean()
               .reset_index(name = "AvgPerformance")
               .sort_values("AvgPerformance", ascending=False)
               .head(1)
)
highAVG_dep_py
```

```python
#Q6/ What Average age of employees for each Department?
avgAGE_Emp_py = (df.groupby("Department")["Age"]
              .mean()
              .reset_index(name = "AVG_Age")
)
avgAGE_Emp_py
```

```python
#Q7/ What Average monthly income by gender ?
avg_income_gender_py = (df.groupby("Gender")["MonthlyIncome"]
                     .mean()
                     .reset_index(name = "AVG_MonthlyIncome")
)
avg_income_gender_py
```

```python
#Q8/ What is the average monthly income by education level?
avg_income_edu_py = (df.groupby("Education")["MonthlyIncome"]
                  .mean()
                  .reset_index(name = "AVG_MonthlyIncome")
                  .sort_values("AVG_MonthlyIncome", ascending=False)
)
avg_income_edu_py
```

```python
#Q9/ Which job role works the most overtime?
mostOvertime_py = (df[df["OverTime"] == "Yes"]
                .groupby("JobRole")["OverTime"]
                .count()
                .reset_index(name="NumberOfOvertime")
                .sort_values("NumberOfOvertime", ascending=False)
)
mostOvertime_py
```

```python
#Q10/ What Average years at company by department ?
avg_year_byDep_py = (df.groupby("Department")["YearsAtCompany"]
                  .mean()
                  .reset_index(name = "avg_Year")
)
avg_year_byDep_py
```

```python
#Q11/ What Average monthly income by job level ?
avg_income_joblevel_py = (df.groupby("JobLevel")["MonthlyIncome"]
                       .mean()
                       .reset_index(name = "avg_income")
                       .sort_values("avg_income" , ascending= False)
)
avg_income_joblevel_py
```

```python
#Q12/ Performance rating by years at company (loyalty vs performance) ?
per_byYear_py = (df.groupby("YearsAtCompany")["PerformanceRating"]
              .mean()
              .reset_index(name = "avg_performance"))
per_byYear_py
```

```python
# STEP DATABASE - connect to data base
conn = sqlite3.connect("hr.db") 
```

```python
# STEP DATABASE - copy dataframe to data base
df.to_sql("employees", conn, if_exists="replace", index=False)
```

```python
#Q1/ How many total employees are there?
total_employees_sql = pd.read_sql_query("""SELECT COUNT(*) AS ALL_EMPLOYEES 
                       FROM employees ;""",conn)
total_employees_sql
```

```python
#Q2/ What is the employee count for each department?
dept_counts_sql = pd.read_sql_query("""SELECT Department , COUNT(*) AS EMPLOYEES
                       FROM employees 
                       GROUP BY Department;""",conn)
dept_counts_sql
```

```python
#Q3/ What is the average monthly income by job role?
avg_income_byJob_sql = pd.read_sql_query("""SELECT JobRole , AVG(MonthlyIncome) AS AVG_MonthlyIncome
                       FROM employees 
                       GROUP BY JobRole;""",conn)
avg_income_byJob_sql
```
